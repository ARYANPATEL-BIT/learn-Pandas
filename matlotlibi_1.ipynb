```python
sales_data = {
    "Months" : ['Jan', 'Feb', "Mar", "Apr",'jun'],
    'Sales' : [150,200,340,380,440],
    'Sales2' : [100,250,320,400,540]
}
```

```python
pandas_sales = pd.DataFrame(sales_data)
```

```python
plt.plot(pandas_sales["Months"], pandas_sales['Sales'], 'r:', marker = "o", label = "Sales")  # if using label we must use plt.legends in order to tell python to display it
plt.plot(pandas_sales["Months"], pandas_sales['Sales2'], 'g--', marker = ".", label = 'Sales2')  # g is for color and -- is for dash line , - is for normal line , "g " space is for no line and g: is for dotted line
plt.xlabel('Time', size = 15 , color = 'r')  # r for red , g for green
plt.ylabel("Sales", size = 15 ,color = 'orange') 
plt.grid(True)
plt.legend()  # required to show labels
plt.show()
```

```python
plt.plot(pandas_sales["Months"], pandas_sales['Sales'], 'r:', marker = "o", label = "Sales")  # if using label we must use plt.legends in order to tell python to display it
plt.plot(pandas_sales["Months"], pandas_sales['Sales2'], 'g--', marker = ".", label = 'Sales2')  # g is for color and -- is for dash line , - is for normal line , "g " space is for no line and g: is for dotted line
plt.xlabel('Time', size = 15 , color = 'r')  # r for red , g for green
plt.ylabel("Sales", size = 15 ,color = 'orange') 
plt.grid(True)
plt.legend(loc = "best")  # loc to change the location of lables  (right (will be top right), left (will be top left) , lower right , lower left , center , best( to give the best position))
plt.show()
```

```python
plt.plot(pandas_sales["Months"], pandas_sales['Sales'], 'r:', marker = "o", label = "Sales")  # if using label we must use plt.legends in order to tell python to display it
plt.plot(pandas_sales["Months"], pandas_sales['Sales2'], 'g--', marker = ".", label = 'Sales2')  # g is for color and -- is for dash line , - is for normal line , "g " space is for no line and g: is for dotted line
plt.xlabel('Time', size = 15 , color = 'r')  # r for red , g for green
plt.ylabel("Sales", size = 15 ,color = 'orange') 
plt.grid(True)
legend = plt.legend(loc = "best")  # loc to change the location of lables  (right (will be top right), left (will be top left) , lower right , lower left , center , best( to give the best position))
legend.get_frame().set_facecolor('C1')   # C is for color and from 0 to 9 there are several different color
plt.title('Sales Trend')  # title for the graph
plt.show()
```
