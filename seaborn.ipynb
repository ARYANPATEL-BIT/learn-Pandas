```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')

print(df)
```

```python
# creating scatter plot

sns.scatterplot(x = df['total_bill'],y = df['tip'], hue = df['time'])  # hue divides the data into all unique parts.

plt.show()
```

```python

sns.scatterplot(data = df, x= 'total_bill', y= 'tip', hue = 'time', style='time') # hue divides the data into all unique parts.

plt.show()
```

```python

sns.scatterplot(data = df, x= 'total_bill', y= 'tip', hue = 'time', style='day') # hue divides the data into all unique parts.

plt.show()
```

```python

sns.scatterplot(data = df, x= 'total_bill', y= 'tip', hue = 'day', style='day') # hue divides the data into all unique parts.

plt.show()
```

```python
#Histogram

sns.histplot(data=df, x = 'total_bill')

plt.show()
```

```python
sns.histplot(data=df, x='total_bill', y = 'tip')
plt.show()
```

```python

sns.histplot(data=df, x = 'total_bill', kde =True)  #kde = kernel density estimate and it tell us about the skewness of data

plt.show()
```

```python
sns.histplot(data=df, x = 'total_bill', y = 'tip',kde =True)

plt.show()
```

```python
round(df['total_bill'].skew(),2)
```

```python
df['total_bill'].skew().round(2)
```

```python
sns.kdeplot(data = df, x = 'total_bill',hue='time')
plt.show()
```

```python
sns.kdeplot(data = df, x = 'total_bill', hue='time', fill= True)
plt.show()
```

```python
sns.kdeplot(data = df, x = 'total_bill', hue='time', fill= True)
plt.show()
```

```python
## HEATMAP 

iris = sns.load_dataset('iris')

corr = iris.select_dtypes(include='number').corr()

plt.figure(figsize=(5,6))

ax = sns.heatmap(corr, annot = True, cmap='coolwarm', linecolor='black')

ax.set_title('co relation heatmap')

plt.show()
```

```python
iris = sns.load_dataset('iris')

sns.pairplot(iris, hue='species', diag_kind='hist')

plt.plot()
```

```python
iris = sns.load_dataset("iris")

sns.pairplot(iris, hue='species', diag_kind='kde', height=2.5, aspect=1)

plt.show()
```

```python
tips = sns.load_dataset('tips')

sns.pairplot(tips, vars=['total_bill','tip','size'], hue='sex', diag_kind='kde')

plt.show()
```

```python
tips = sns.load_dataset('tips')

sns.boxplot(data=tips, x='day', y='total_bill')

plt.show()
```

```python
#Heatmap and Pivot table


pivot = df.pivot_table(index='day', columns='time',values='total_bill',aggfunc='mean')
sns.heatmap(pivot, annot=True, fmt='.2f', cmap='coolwarm')
plt.show()
```
