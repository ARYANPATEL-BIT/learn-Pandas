```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')

print(df)
```

```python
# creating scatter plot

sns.scatterplot(x = df['total_bill'],y = df['tip'], hue = df['time'])  # hue divides the data into all unique parts.

plt.show()
```

```python

sns.scatterplot(data = df, x= 'total_bill', y= 'tip', hue = 'time', style='time') # hue divides the data into all unique parts.

plt.show()
```

```python

sns.scatterplot(data = df, x= 'total_bill', y= 'tip', hue = 'time', style='day') # hue divides the data into all unique parts.

plt.show()
```

```python
#Histogram

sns.histplot(data=df, x = 'total_bill')

plt.show()
```

```python
sns.histplot(data=df, x='total_bill', y = 'tip')
plt.show()
```

```python

sns.histplot(data=df, x = 'total_bill', kde =True)  #kde = kernel density estimate and it tell us about the skewness of data

plt.show()
```

```python
sns.histplot(data=df, x = 'total_bill', y = 'tip',kde =True)

plt.show()
```

```python
round(df['total_bill'].skew(),2)
```

```python
df['total_bill'].skew().round(2)
```

```python
sns.kdeplot(data = df, x = 'total_bill', hue='time')
plt.show()
```

```python
sns.kdeplot(data = df, x = 'total_bill', hue='time', fill= True)
plt.show()
```

```python
sns.kdeplot(data = df, x = 'total_bill', hue='time', fill= True)
plt.show()
```
